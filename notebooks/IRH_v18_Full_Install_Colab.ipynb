## üéõÔ∏è Interactive Testing Menu

Select the scale and level of testing you wish to run.

```python
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import time

# Test configuration
test_config = {
    'level': 'quick',
    'modules': []
}

# Header
header = widgets.HTML(
    value="""
    <div style='background: linear-gradient(135deg, #1a1a2e 0%, #16213e 100%); 
                padding: 20px; border-radius: 10px; margin-bottom: 20px;'>
        <h2 style='color: #00d4ff; margin: 0;'>üéõÔ∏è IRH v18.0 Testing Menu</h2>
        <p style='color: #a0a0a0; margin: 5px 0 0 0;'>Select test level and modules to validate</p>
    </div>
    """
)

# Test level selector
level_selector = widgets.RadioButtons(
    options=[
        ('üöÄ Quick (~30 sec) - Core verification', 'quick'),
        ('üìä Standard (~2 min) - All modules', 'standard'),
        ('üî¨ Comprehensive (~5 min) - Full validation', 'comprehensive'),
        ('üß™ Full Test Suite (~10 min) - pytest execution', 'pytest')
    ],
    value='quick',
    description='Test Level:',
    style={'description_width': '100px'},
    layout=widgets.Layout(width='100%')
)

# Module selection
module_selector = widgets.SelectMultiple(
    options=[
        ('üåê Standard Model Topology', 'topology'),
        ('üîÑ RG Flow & Fixed Point', 'rg_flow'),
        ('‚ö° Electroweak Sector', 'electroweak'),
        ('üî¨ Neutrino Physics', 'neutrino'),
        ('üåå Dark Energy', 'dark_energy'),
        ('üé≠ Quantum Mechanics', 'quantum'),
        ('üìê Emergent Gravity', 'gravity'),
        ('‚è±Ô∏è Emergent Spacetime', 'spacetime'),
        ('üîÆ Complete QFT', 'qft'),
        ('üß≤ Strong CP Problem', 'strong_cp'),
        ('üé® Flavor Mixing', 'flavor')
    ],
    value=['topology', 'rg_flow', 'qft'],
    description='Modules:',
    style={'description_width': '100px'},
    layout=widgets.Layout(width='100%', height='200px')
)

# Output area
output_area = widgets.Output()

# Run button
run_button = widgets.Button(
    description='üöÄ Run Tests',
    button_style='success',
    layout=widgets.Layout(width='200px', height='40px')
)

def run_tests(button):
    with output_area:
        clear_output()
        level = level_selector.value
        modules = list(module_selector.value)
        
        print("="*60)
        print(f"Running {level.upper()} tests on {len(modules)} module(s)...")
        print("="*60 + "\n")
        
        if level == 'pytest':
            print("Executing pytest suite...\n")
            !cd python && export PYTHONPATH=$(pwd)/src && python -m pytest tests/v18/ -v --tb=short 2>&1 | head -100
        else:
            run_validation(level, modules)

run_button.on_click(run_tests)

# Display menu
display(header)
display(widgets.VBox([
    widgets.HTML("<h4>Test Configuration</h4>"),
    level_selector,
    widgets.HTML("<br><h4>Select Modules (Ctrl/Cmd+Click for multiple)</h4>"),
    module_selector,
    widgets.HTML("<br>"),
    run_button,
    output_area
]))
```

```python
# Test execution functions
import numpy as np
from tqdm.notebook import tqdm

def run_validation(level, modules):
    """Run validation tests based on selected level and modules."""
    results = {}
    start_time = time.time()
    
    # Define test functions for each module
    tests = {
        'topology': test_topology,
        'rg_flow': test_rg_flow,
        'electroweak': test_electroweak,
        'neutrino': test_neutrino,
        'dark_energy': test_dark_energy,
        'quantum': test_quantum,
        'gravity': test_gravity,
        'spacetime': test_spacetime,
        'qft': test_qft,
        'strong_cp': test_strong_cp,
        'flavor': test_flavor
    }
    
    # Run selected tests
    for module in tqdm(modules, desc="Testing modules"):
        if module in tests:
            try:
                results[module] = tests[module](level)
            except Exception as e:
                results[module] = {'passed': False, 'error': str(e)}
    
    # Summary
    elapsed = time.time() - start_time
    passed = sum(1 for r in results.values() if r.get('passed', False))
    
    print("\n" + "="*60)
    print(f"üìä Test Summary")
    print("="*60)
    print(f"Level: {level}")
    print(f"Modules tested: {len(modules)}")
    print(f"Passed: {passed}/{len(modules)}")
    print(f"Time: {elapsed:.2f}s")
    print("="*60)
    
    for module, result in results.items():
        status = "‚úÖ" if result.get('passed') else "‚ùå"
        print(f"  {status} {module}: {result.get('message', result.get('error', 'Unknown'))}")
    
    return results

def test_topology(level):
    """Test Standard Model topology derivation."""
    from irh.core.v18 import StandardModelTopology
    sm = StandardModelTopology()
    result = sm.compute_full_derivation()
    
    checks = {
        'beta_1': result['gauge_sector']['beta_1'] == 12,
        'n_inst': result['matter_sector']['n_inst'] == 3,
        'verified': sm.verify_standard_model()
    }
    
    passed = all(checks.values())
    return {
        'passed': passed,
        'message': f"Œ≤‚ÇÅ={result['gauge_sector']['beta_1']}, n_inst={result['matter_sector']['n_inst']}",
        'details': checks
    }

def test_rg_flow(level):
    """Test RG flow and fixed point."""
    from irh.core.v18 import find_fixed_point, BetaFunctions
    
    fp = find_fixed_point()
    beta = BetaFunctions()
    b_vals = beta.evaluate(fp.lambda_star, fp.gamma_star, fp.mu_star)
    
    # Theoretical values
    lambda_theory = 48 * np.pi**2 / 9
    gamma_theory = 32 * np.pi**2 / 3
    mu_theory = 16 * np.pi**2
    
    checks = {
        'lambda': np.isclose(fp.lambda_star, lambda_theory, rtol=1e-6),
        'gamma': np.isclose(fp.gamma_star, gamma_theory, rtol=1e-6),
        'mu': np.isclose(fp.mu_star, mu_theory, rtol=1e-6),
        'beta_zero': all(abs(b) < 1e-6 for b in b_vals)
    }
    
    return {
        'passed': all(checks.values()),
        'message': f"Œª*={fp.lambda_star:.4f}, Œ≥*={fp.gamma_star:.4f}, Œº*={fp.mu_star:.4f}",
        'details': checks
    }

def test_electroweak(level):
    """Test electroweak sector predictions."""
    from irh.core.v18 import ElectroweakSector
    
    ew = ElectroweakSector()
    sector = ew.compute_full_sector()
    
    checks = {
        'higgs_mass': 120 < sector['higgs']['mass'] < 130,  # ~125 GeV
        'weinberg': 0.20 < sector['weinberg_angle']['sin2_theta_w'] < 0.25,
        'w_mass': 75 < sector['gauge_bosons']['w_mass'] < 85,
        'z_mass': 86 < sector['gauge_bosons']['z_mass'] < 96
    }
    
    return {
        'passed': all(checks.values()),
        'message': f"m_H={sector['higgs']['mass']:.1f}GeV, sin¬≤Œ∏_W={sector['weinberg_angle']['sin2_theta_w']:.4f}",
        'details': checks
    }

def test_neutrino(level):
    """Test neutrino sector predictions."""
    from irh.core.v18 import NeutrinoSector
    
    neutrino = NeutrinoSector()
    hierarchy = neutrino.compute_mass_hierarchy()
    masses = neutrino.compute_absolute_masses()
    
    checks = {
        'hierarchy': hierarchy['hierarchy'] == 'normal',
        'sum_bound': masses['sum_masses_eV'] < 0.12,
        'positive': all(m > 0 for m in masses['masses_eV'])
    }
    
    return {
        'passed': all(checks.values()),
        'message': f"Hierarchy: {hierarchy['hierarchy']}, Œ£mŒΩ={masses['sum_masses_eV']:.4f}eV",
        'details': checks
    }

def test_dark_energy(level):
    """Test dark energy predictions."""
    from irh.core.v18 import DarkEnergyModule
    
    de = DarkEnergyModule()
    analysis = de.compute_full_analysis()
    w0 = analysis['equation_of_state']['w0']
    
    checks = {
        'w0_range': -1.1 < w0 < -0.9,  # Near phantom threshold
        'de_positive': analysis['vacuum_energy']['rho_de'] > 0
    }
    
    return {
        'passed': all(checks.values()),
        'message': f"w‚ÇÄ={w0:.4f}",
        'details': checks
    }

def test_quantum(level):
    """Test emergent quantum mechanics."""
    from irh.core.v18 import EmergentQuantumMechanics
    
    qm = EmergentQuantumMechanics()
    summary = qm.get_summary()
    
    checks = {
        'born_rule': summary['born_rule']['derived'],
        'unitarity': summary['unitarity_preserved']
    }
    
    return {
        'passed': all(checks.values()),
        'message': f"Born rule derived: {checks['born_rule']}, Unitary: {checks['unitarity']}",
        'details': checks
    }

def test_gravity(level):
    """Test emergent gravity."""
    from irh.core.v18 import compute_emergent_gravity_summary
    
    gravity = compute_emergent_gravity_summary()
    
    checks = {
        'einstein': gravity['einstein_equations']['derived'],
        'graviton': gravity['graviton_propagator']['massless']
    }
    
    return {
        'passed': all(checks.values()),
        'message': f"Einstein eqs: {checks['einstein']}, Massless graviton: {checks['graviton']}",
        'details': checks
    }

def test_spacetime(level):
    """Test emergent spacetime."""
    from irh.core.v18 import EmergentSpacetime
    
    st = EmergentSpacetime()
    props = st.verify_all_properties()
    
    return {
        'passed': all(props.values()),
        'message': f"Lorentzian: {props['lorentzian_signature']}, 4D: {props['four_dimensional']}",
        'details': props
    }

def test_qft(level):
    """Test complete QFT emergence."""
    from irh.core.v18 import EmergentQFT
    
    qft = EmergentQFT()
    verified = qft.verify_standard_model()
    
    return {
        'passed': all(verified.values()),
        'message': f"All SM features: {all(verified.values())}",
        'details': verified
    }

def test_strong_cp(level):
    """Test strong CP resolution."""
    from irh.core.v18 import StrongCPResolution
    
    cp = StrongCPResolution()
    resolution = cp.verify_resolution()
    
    checks = {
        'theta_zero': resolution['theta_effective'] == 0,
        'resolved': resolution['resolved']
    }
    
    return {
        'passed': all(checks.values()),
        'message': f"Œ∏={resolution['theta_effective']}, Resolved: {resolution['resolved']}",
        'details': checks
    }

def test_flavor(level):
    """Test flavor mixing matrices."""
    from irh.core.v18 import CKMMatrix, PMNSMatrix
    
    ckm = CKMMatrix()
    pmns = PMNSMatrix()
    
    ckm_unitary = ckm.verify_unitarity()
    pmns_unitary = pmns.verify_unitarity()
    
    checks = {
        'ckm_unitary': ckm_unitary['is_unitary'],
        'pmns_unitary': pmns_unitary['is_unitary']
    }
    
    return {
        'passed': all(checks.values()),
        'message': f"CKM unitary: {checks['ckm_unitary']}, PMNS unitary: {checks['pmns_unitary']}",
        'details': checks
    }

print("‚úÖ Test functions loaded. Use the menu above to run tests.")
```

## üìà Detailed Results Visualization

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_predictions():
    """Visualize IRH predictions vs experimental data."""
    from irh.core.v18 import (
        FineStructureConstant, ElectroweakSector,
        NeutrinoSector, DarkEnergyModule
    )
    
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    fig.suptitle('IRH v18.0 Predictions vs Experiment', fontsize=14, fontweight='bold')
    
    # 1. Fine structure constant
    ax = axes[0, 0]
    alpha = FineStructureConstant()
    irh_alpha = alpha.compute_alpha_inverse()['alpha_inverse']
    exp_alpha = 137.035999084
    ax.barh(['IRH', 'Experiment'], [irh_alpha, exp_alpha], color=['#2ecc71', '#3498db'])
    ax.set_xlabel('Œ±‚Åª¬π')
    ax.set_title('Fine Structure Constant')
    ax.set_xlim(137.03, 137.04)
    
    # 2. Electroweak masses
    ax = axes[0, 1]
    ew = ElectroweakSector()
    sector = ew.compute_full_sector()
    particles = ['Higgs', 'W', 'Z']
    irh_masses = [sector['higgs']['mass'], 
                  sector['gauge_bosons']['w_mass'],
                  sector['gauge_bosons']['z_mass']]
    exp_masses = [125.25, 80.4, 91.2]
    x = np.arange(len(particles))
    width = 0.35
    ax.bar(x - width/2, irh_masses, width, label='IRH', color='#2ecc71')
    ax.bar(x + width/2, exp_masses, width, label='Experiment', color='#3498db')
    ax.set_xticks(x)
    ax.set_xticklabels(particles)
    ax.set_ylabel('Mass (GeV)')
    ax.set_title('Electroweak Masses')
    ax.legend()
    
    # 3. Neutrino masses
    ax = axes[1, 0]
    neutrino = NeutrinoSector()
    masses = neutrino.compute_absolute_masses()
    ax.bar(['ŒΩ‚ÇÅ', 'ŒΩ‚ÇÇ', 'ŒΩ‚ÇÉ'], [m*1000 for m in masses['masses_eV']], color='#9b59b6')
    ax.axhline(y=0.12*1000/3, color='r', linestyle='--', label='Bound (Œ£mŒΩ < 0.12 eV)/3')
    ax.set_ylabel('Mass (meV)')
    ax.set_title(f"Neutrino Masses (Œ£mŒΩ = {masses['sum_masses_eV']*1000:.1f} meV)")
    ax.legend()
    
    # 4. Dark energy w0
    ax = axes[1, 1]
    de = DarkEnergyModule()
    w0 = de.compute_full_analysis()['equation_of_state']['w0']
    w_range = np.linspace(-1.2, -0.8, 100)
    ax.axvline(x=w0, color='#2ecc71', linewidth=3, label=f'IRH: w‚ÇÄ = {w0:.4f}')
    ax.axvline(x=-1, color='gray', linestyle='--', label='Cosmological constant')
    ax.axvspan(-0.89, -0.76, alpha=0.3, color='blue', label='DESI (2024)')
    ax.set_xlim(-1.15, -0.75)
    ax.set_xlabel('w‚ÇÄ')
    ax.set_title('Dark Energy Equation of State')
    ax.legend()
    ax.set_yticks([])
    
    plt.tight_layout()
    plt.show()

plot_predictions()
```
